```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

df=pd.read_excel('mlpaper.xlsx')

X = df.drop(columns = ['power'])

y = df['power']
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.15,random_state=42)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaler = scaler.fit_transform(X_train)
X_test_scaler = scaler.fit_transform(X_test)

from sklearn.svm import SVR
svr = SVR().fit(X_train_scaler,y_train)

svr1 = SVR(kernel = 'rbf',gamma = 'scale',degree=6).fit(X_train_scaler,y_train)
y_pred1 = svr1.predict(X_test_scaler)
```

```python
plt.scatter(y_test, y_pred1)
plt.xlabel('Actual Power')
plt.ylabel('Predicted Power')
plt.title('Actual vs Predicted Power')
plt.show()
```

```python
residuals = y_test - y_pred1
plt.scatter(y_pred1, residuals)
plt.xlabel('Predicted Power')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```
